Установка датасета lenta.ru. Для данной задачи берем только наиболее новые title, начиная с 2018

```python
# Download latest version
lenta_path = kagglehub.dataset_download("yutkin/corpus-of-russian-news-articles-from-lenta")

print("Path to dataset files:", lenta_path)
```

```python
lenta_df = pd.read_csv(f'{lenta_path}/lenta-ru-news.csv')
```

```python
lenta_df = lenta_df.iloc[700000:800974].reset_index(drop=True)
lenta_df = lenta_df.drop(['url', 'topic', 'tags', 'date'], axis = 1)
```

Убираем ненужные спецсимволы из столбца title, например \xa0 для неразрывного пробела

```python
lenta_df["title"] = (
    lenta_df["title"]
    .str.replace("\xa0", " ", regex=False)
    .str.replace("\n", " ", regex=False)
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
)
```

```python
poems_path = '/content/poems_n_Koziev_data.txt'
synthetic_data_path = '/content/synthetic_data.txt'
```

```python
def create_DataFrame_from_txt(file_path, column_name="text"):
    """
    Читает текстовый файл и возвращает DataFrame со столбцом text.

    Параметры:
    - file_path: путь к .txt файлу
    - column_name: имя колонки в DataFrame

    Возвращает:
    - pd.DataFrame
    """
    with open(file_path, "r", encoding="utf-8") as file:
        lines = file.readlines()

    lines = [line.strip() for line in lines if line.strip()]
    df = pd.DataFrame(lines, columns=[column_name])

    return df
```

```python
synthetic_df = create_DataFrame_from_txt(synthetic_data_path)
poems_df = create_DataFrame_from_txt(poems_path)
```

Объединяем все данные в 1 датасет

```python
text_from_lenta = lenta_df["title"].astype(str)
text_from_synth = synthetic_df["text"].astype(str)
text_from_poems = poems_df["text"].astype(str)

combined_series = pd.concat([text_from_synth, text_from_poems, text_from_lenta], ignore_index=True)


df_combined = pd.DataFrame({"text": combined_series})
```

```python
df_combined.tail()
```

Удаляем ненужные символы \/ в тексте. Для примеров вида: "продам б/у утюг" -> "продам бу утюг", как и в тестовых данных. Перемешиваем данны, а также создаем новый столбец без пробелов

```python
df_combined["text"] = df_combined["text"].str.replace(r"[\\/]", "", regex=True)
df_shuffled = df_combined.sample(frac=1, random_state=42).reset_index(drop=True)
df_shuffled["no_space"] = df_shuffled["text"].str.replace(" ", "", regex=False)
```

Сохраняем готовый датасет в csv

```python
df_shuffled.to_csv("text_data.csv", index=False)
```
